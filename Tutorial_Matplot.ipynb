-------------
- **The basic steps to creating plots with matplotlib are:**
    - Prepare the Data
    - Create a plot
    - Plot
    - Customize plot
    - Save plot
    - Show plot
    
  --------

## 1. Prepare the Data:

```python
import numpy as np

# 1-D data
x=np.linspace(0,10,100)
y=np.cos(x)
z=np.sin(x)

# 2-D data 
data= 2 * np.random.random((10,10))
data2= 3 * np.random.random((10,10))
Y, X= np.mgrid[-3:3:100j, -3:3:100j]
U= -1 -X**2 + Y
V= 1 + X - Y**2

# Image data
from matplotlib.cbook import get_sample_data
img= np.load(get_sample_data('axes_grid/bivariate_normal.npy'))
```

## 2. Create Plot:

```python
import matplotlib.pyplot as plt
```

### Figure

```python

fig= plt.figure()
fig2= plt.figure(figsize=plt.figaspect(2.0))
```

### Axes

**Axes:** All plotting is done with respect to an $Axes$. In most cases, a subplot will fit your needs. A subplot is an axes on a grid system

```python

fig.add_axes()
ax1= fig.add_subplot(221)  # row-col-num
ax3= fig.add_subplot(212)

fig3, axes= plt.subplots(nrows=2, ncols=2)
fig4, axes2= plt.subplots(ncols=3)
```

-----------
## 3. Plotting Routines

### 1-D data

```python

fig,ax= plt.subplots()

lines= ax.plot(x,y) # Draws a point, with lines or markers connecting them

ax.scatter(x,y,color='k',alpha=0.5) # Draws unconnected points, scaled or colored

ax.fill(x,y,color='yellow') # Draw a filled polygons

ax.fill_between(x,y,color='red') # Fill between y-values and origin
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2)

axes[0,0].bar([1,2,3],[3,4,5]) # Plots vertical rectangles (constant width)

axes[1,0].barh([0.5,1,2.5],[0,1,2]) # Plot horizontal rectangles (const. ht)

axes[1,1].axhline(0.45) # Draw a horizontal line across axes

axes[0,1].axvline(0.65) # Draw a vertical line across axes
```

### 2D data or Images

```python
# 2-D data or Images
fig, axes2= plt.subplots(nrows=1,ncols=3)

im= ax.imshow(img,cmap='gist_earth', interpolation='nearest'
              ,vmin=-3,vmax=3)  # Colormapped or RGB arrays 

axes2[0].pcolor(data2) # Pseudocolor plot of 2D array

axes2[0].pcolormesh(data) # Pseudocolor plot of 2D array

CS= plt.contourf(Y,X,U) # Plot contours

axes2[1].contourf(V) # Plot filled contours

axes2[1]=ax.clabel(CS) # Label a contour plot

axes2[2].contourf(U) # Plot filled contours

axes2[2]=ax.clabel(CS) #  Label a contour plot
```

### Vector Fields

```python
# Vector Fields
fig,axes= plt.subplots(nrows=1,ncols=3)

axes[0].arrow(0,0,0.5,0.5) # Adds an arrow to the axes

axes[1].quiver(y,z) # Plots a 2D field of arrows

axes[2].streamplot(X,Y,U,V) # Plots a 2D field of arrows
```

### Data Distributions

```python
# Data distributions
fig,axes=plt.subplots(nrows=1,ncols=3)

axes[0].hist(y) # Plots a histogram of passed data

axes[1].boxplot(y) # Makes a box and whisker plot

axes[2].violinplot(z) # Makes a violin plot
```

-----------
## 4. Customize Plot:


### Colors, Color Bars and Color Maps

```python
# Colors, Color Bars and Color Maps

fig, axes=plt.subplots(nrows=2, ncols=1)

axes[0].plot(x,x,  x,x**2,  x,x**3)  # Plotting x vs x; x vs x**2 and x vs x**3


axes[1].plot(x,y,c='k')      # Plotting x vs y; with color 'c' parameter,  
                             # controlling color
    
    
axes[1].plot(x,y, alpha=0.4) # Plotting x vs y; with alpha parameter, 
                             # controlling color intensity
    
    
```

### Image Data

```python
# Image Data
fig,ax= plt.subplots()
im= ax.imshow(img,cmap='gist_earth', interpolation='nearest',
             vmin= -2, vmax=2)  # Colormapped or RGB arrays

fig.colorbar(im, orientation='horizontal')

im= ax.imshow(img, cmap='seismic')
```

###  Markers

```python
# Markers
fig, ax = plt.subplots(nrows=3,ncols=1)

ax[0].plot(x,y,c='k')     # Creating a original plot,

ax[1].scatter(x,y,marker=".")

ax[2].plot(x,y,marker='o', alpha=0.2)
```

### Linestyles

```python
# Linestyles

fig,axes=plt.subplots(nrows=2, ncols=2)

axes[0,0].plot(x,y,linewidth=4.0)

axes[0,1].plot(x,y,ls='solid')

axes[1,0].plot(x,y,ls='--')

axes[1,1].plot(x,y,'--',  x**2,y**2, '-.')

```

### Text and Annotations

```python
# Text and Annotations

fig,ax=plt.subplots()

ax.plot(x,y)

ax.text(2,1.25,'Example Graph', style='italic', fontsize=20)

ax.annotate("Sine", xy=(8,0), xycoords='data',xytext=(11.5,0),
           textcoords='data',
           arrowprops=dict(arrowstyle="->", connectionstyle='arc3'))
```

### Math text

```python
# Math text

plt.title(r'$sigma_i=15$', fontsize=20)
```

### Limits, Legends and Layouts

```python
# Limits and Autoscaling

fig,ax=plt.subplots(nrows=2,ncols=2)

ax[0,0].plot(x,y)

ax[0,0].text(0,1.25,'Original Graph', style='italic', fontsize=15)
 

ax[0,1].margins(x=0.0, y=0.1)  # Adds padding to a plot

ax[0,1].set(xlim=[0,10.5], ylim=[-1.5,1.5]) # Set limits for x- and y- axis

ax[0,1].text(0,1.70,'Limits for x- and y- ', style='italic', fontsize=15)


ax[1,0].set_xlim(0,10.5)  # Sets limit for x- axis only

ax[1,0].text(0,-0.35,'Limits for x- only', style='italic', fontsize=15)


ax[1,1].axis('equal')  # Set the aspect ratio of the plot to 1

ax[1,1].text(-0.055,-0.060,'Aspect to eqaul', style='italic', fontsize=15)
```

### Legends

```python
# Legends
fig,axes= plt.subplots()

axes.plot(x,y)

axes.set(title='An Example Plot', ylabel='Y-Axis', xlabel='X- Axis')
# Sets a title and x- and y- axis labels


axes.legend(loc='best')
# No overlapping plot elements
```

### Ticks

```python
fig, axes= plt.subplots(nrows=2,ncols=1)

axes[0].plot(x,y)
axes[1].plot(x,y)

axes[0].xaxis.set(ticks=range(1,5), ticklabels=[3,100,-12,"foo"])
# Manually set x- ticks

axes[1].tick_params(axis='y', direction='inout', length=10)
# Makes y- ticks longer and go in and out of the axis




axes[1].annotate("Tick", xy=(-0.50,1), xycoords='data',xytext=(-2,0),
           textcoords='data',
           arrowprops=dict(arrowstyle="->", connectionstyle='arc3'))
```

### Subplot Spacing:

```python
# Subplot Spacing
fig, axes= plt.subplots(nrows=2,ncols=2)

axes[0,0].plot(x,x)
axes[0,1].plot(x,x**2)
axes[1,0].plot(x,y)
axes[1,1].plot(x,x**3)

fig.subplots_adjust(wspace=0.1,   # left,bottom,top, right acts as coordinates
                    hspace=0.5,   # wspace, thickness of vertical separation
                    left=0.0,     # hspace, thickness of horizontal separation
                    right=10,
                    top=10.0,   
                    bottom=5.0)
# Adjusts the spacing between subplots

fig.tight_layout()
# Fit subplots into the figure area



'''Comment out , fig.tight_layout() command to see the effects of 
previous commands'''
```

### Axis Spines:


```python
# Axis Spines
fig, axes= plt.subplots(nrows=1,ncols=3)

axes[0].plot(x,y)

axes[1].plot(x,y)
axes[1].spines['top'].set_visible(False) # Making top axis line, invisible

axes[2].plot(x,y)
axes[2].spines['bottom'].set_position(('outward',10)) # Moving the bottom 
                                                      # axis line outward
fig.tight_layout() 
```

## 5. Saving Plot:


```python
# Save figures
plt.savefig('foo.png')

# Saving transparent figures
plt.savefig('foo.png', transparent=True)
```

## 6. Show Plot

```python
# Show plot
plt.show()

# Close and Clear
plt.cla()  # Clear an axis

plt.clf()  # Clear the entire figure

plt.close() # Close a window
```

## WorkFlow:

```python
import matplotlib.pyplot as plt

# Step 1
x=[1,2,3,4]
y=[10,20,30,40]   


# Step 2
fig=plt.figure()


# Step 3
ax=fig.add_subplot(111)


# Step 4
ax.plot(x, y, color='lightblue', linewidth=3)

ax.scatter([2,4,6],
          [5,15,25],
          color='darkgreen',
          marker='^')

ax.set_xlim(1,6.5)


# Step 5
plt.savefig('foo.png')

# Step 6
plt.show()
```
